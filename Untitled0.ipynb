```python
def posneg(x):
  if x<0:
    return 'il num è neg'
  elif x==0:
    return 'il num inserito è zero, cambiare num'
  else:
    return 'il num è pos'
```

```python
posneg(1)
```

```python
def paridispari(x):
  if x%2==0:
    return 'il num è pari'
  else:
    return 'il num è dispari'
```
